```python
# return_dict_train['sig_all']
# write a function for gaussian distribution given mean and variance
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.))) / (np.sqrt(2*np.pi) * sig)

```
